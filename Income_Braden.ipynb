## Why Income?


-Usually less crime in wealthier neighborhoods
-Better public service, usally more local stores
-Easier to finance houses
-Houses in wealthier neighborhood will gain more value over time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("income.csv")
data = df.sort_values(by="2009 Median Income ('09 Dollars)", ascending=True)
data = data.iloc[70:] 
categories = data['Neighborhood']
income = data["2009 Median Income ('09 Dollars)"]

plt.barh(categories, income)
plt.xlabel('Income')
plt.ylabel('')
plt.title('Neighborhoods and Their Median Income (2009)')

plt.show()
```
